###**Gerador Datasets Artificiais**

```python
import numpy as np
from sklearn.datasets import make_blobs, make_moons
import matplotlib.pyplot as plt
```

```python
# Gerar um único dataset combinando dois globulares e uma estrutura de lua
np.random.seed(42)

# Dados globulares
globular1, _ = make_blobs(n_samples=150, centers=2, cluster_std=0.50, random_state=0)
globular2, _ = make_blobs(n_samples=150, centers=2, cluster_std=0.50, random_state=1)

# Dados aninhados (em forma de luas)
nested, _ = make_moons(n_samples=500, noise=0.07)

# Combinando todos os dados
combined_dataset = np.vstack([globular1, globular2 * np.array([1, 2.5]) + 3, nested * np.array([2, 0.5]) + [5, -3]])

# Plotando o dataset combinado
plt.figure(figsize=(8, 6))
plt.scatter(combined_dataset[:, 0], combined_dataset[:, 1], alpha=0.7)
plt.title("Dataset 01")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons

# Configurando o random seed para consistência
np.random.seed(0)

# Gerando dados globulares com uma distribuição maior
globular3_large, _ = make_blobs(n_samples=150, centers=1, cluster_std=1.0, random_state=2)
globular4_large, _ = make_blobs(n_samples=150, centers=1, cluster_std=1.0, random_state=3)

# Gerando uma lua maior
nested_larger, _ = make_moons(n_samples=300, noise=0.02)
nested_larger_scaled = nested_larger * np.array([2.5, 1.5])

# Definindo um traço linear menor
x_line_short = np.linspace(-5, 5, 100)
y_line_short = 2 * x_line_short + 3
linear_short = np.column_stack((x_line_short, y_line_short))

# Combinando todos os elementos no dataset final
combined_dataset_2_final = np.vstack([
    globular3_large + [20, 10],
    globular4_large * np.array([0.5, 2]) + [30, -10],
    nested_larger_scaled - [20, 10],
    linear_short  # Usando o traço menor definido anteriormente
])

# Plotando o dataset final com a lua aumentada
plt.figure(figsize=(8, 6))
plt.scatter(combined_dataset_2_final[:, 0], combined_dataset_2_final[:, 1], alpha=0.7)
plt.title("Dataset 02")
plt.show()
```

```python
# Gerar um dataset com 1 globular, uma meia lua e uma estrutura que favoreça clusters hierárquicos
np.random.seed(10)

# Dados globulares
globular_single, _ = make_blobs(n_samples=200, centers=1, cluster_std=0.5, random_state=1)

# Dados aninhados (em forma de meia lua)
half_moon, _ = make_moons(n_samples=250, noise=0.05)
half_moon = half_moon[half_moon[:, 1] > 0]  # Selecionar apenas a parte superior da lua

# Estrutura para favorecer clusters hierárquicos
X_hier, _ = make_blobs(n_samples=150, centers=1, cluster_std=0.1)
hierarchical = np.vstack([X_hier * 0.5 + np.array([3, 2]), X_hier + np.array([3, 4])])

# Combinando todos os dados
combined_dataset_3 = np.vstack([globular_single, half_moon * np.array([1.5, 1.5]) + [3, 1], hierarchical])

# Plotando o dataset
plt.figure(figsize=(8, 6))
plt.scatter(combined_dataset_3[:, 0], combined_dataset_3[:, 1], alpha=0.7)
plt.title("Dataset 03")
plt.show()
```
